```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
pd.options.display.max_columns = 999
```

```python
discover_data = pd.read_csv('discover_print_to_csv_2.csv')
```

```python
email_skin_concerns = discover_data[['email', 'request.concerns', 'request.gender', 'request.ageRange', 'request.sensitivity', 'request.skinType']].copy().drop_duplicates()
```

```python
# Convert string representation of list to list
import ast

email_skin_concerns['request.concerns'] = email_skin_concerns['request.concerns'].apply(ast.literal_eval)
```

```python
def remove_skin_types(x):
    terms_to_delete = ['Sensitivity', 'Combination','Dry to Very Dry', 'Normal', 'Oily']
    replacement_list = []
    for entry in x:
        if entry in terms_to_delete:
            del entry
        else:
            replacement_list.append(entry)
    return replacement_list
email_skin_concerns['request.concerns'] = email_skin_concerns['request.concerns'].map(remove_skin_types)
```

```python
email_skin_concerns_expanded = pd.concat([email_skin_concerns, pd.DataFrame(email_skin_concerns['request.concerns'].apply(pd.Series))], axis=1).drop('request.concerns', axis=1)
```

```python
email_skin_concerns_expanded.columns = ['email','request.gender','request.ageRange','request.sensitivity','request.skinType',
                                             'concern1','concern2','concern3','concern4','concern5','concern6','concern7',
                                             'concern8','concern9','concern10','concern11','concern12']
```

```python
email_skin_concerns_expanded = email_skin_concerns_expanded.set_index(['email', 'request.gender', 'request.ageRange', 'request.sensitivity',
       'request.skinType']).stack().reset_index()
```

```python
email_profile_complete = pd.concat([email_skin_concerns_expanded, pd.get_dummies(email_skin_concerns_expanded[0])] ,axis=1).drop(['level_5',0], axis = 1)
```

```python
email_skin_concerns = email_skin_concerns.drop_duplicates(subset=['email'])
email_skin_concerns = pd.concat([email_skin_concerns,pd.get_dummies(email_skin_concerns['request.concerns'].apply(pd.Series).stack()).sum(level=0) ], axis = 1)
```

```python
email_skin_concerns.drop('request.concerns', axis=1, inplace=True)
```
